### MESSAGES

Fundamental unit of context for models in langchain. They represent the input and output of models, carrying both the content and the metadata needed to represent the state of a conversation when interacting with an LLM. they are object that contain:



1: ROLE- identifies the type of message(system,user,ai etc).

2: Content- Represents the actual content of the message (like text,image,audio,document etc).

3: MetaData- Optional fields such as response information, message IDs, and token usage.


Langchain provides a standard message type that works for all model providers, ensuring consistent behaviour regardless of the model called

```python
import os 
from langchain.chat_models import init_chat_model
os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")
model=init_chat_model("groq:llama-3.3-70b-versatile")
```

```python
response = model.invoke("what is langchain")
response.content
```

## Types of message

1: System message: tells the model how to behave and provide context for instructions.

2: Human message: Represents the user input and interactions with the model.

3:AI message: Responses generated by the model, including text content, tool calls, and metadata.

4:Tool Message: represents the outputs of tool calls.

```python
from langchain.messages import SystemMessage,HumanMessage,AIMessage,ToolMessage

message=[
    SystemMessage("You are a comedian who adds puns with each response"),
    HumanMessage("What is artificial intelligence")
]
response=model.invoke(message)
response.content
```

```python
sys_message = SystemMessage("You are a recruiter at cisco you are about to interview a 4th year student from jaypee institute of information technology")
messages=[sys_message,HumanMessage("What are the question do you plan to ask the candidate during the interview process")]

response=model.invoke(messages)

response.content
```

```python
#With a human message we can essentially add three things role,content and metadata.
human_msg=HumanMessage(
    content="Hello!",
    name="Alice",
    id="msg_123"
)
response=model.invoke([human_msg])
response
```

```python
messages=[SystemMessage("You are an astrophysist"),HumanMessage("What is the largest blackhole ever discovered"),
AIMessage("I'd be glad to help you :)  !!!!"),
HumanMessage("Great! what is kardeshev scale")
]
response=model.invoke(messages)
print(response.content)
```

```python
response.usage_metadata
```

#### TOOL MESSAGE

```python
from langchain_core.messages import (
    HumanMessage,
    AIMessage,
    ToolMessage
)

messages = [
    HumanMessage(content="What's the weather in California?"),

    AIMessage(
        content="",
        tool_calls=[
            {
                "name": "get_weather",
                "args": {"location": "California"},
                "id": "call123"
            }
        ]
    ),

    ToolMessage(
        content="Sunny, 72°F",
        tool_call_id="call123"
    )
]

response=model.invoke(messages)
print(response.content)
```
